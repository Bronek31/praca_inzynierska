```python
import requests
from bs4 import BeautifulSoup
import os
import re

def clean_folder_name(folder_name):
    return re.sub(r'[\/:*?"<>|]', '', folder_name)

download_directory = r"C:\Users\48516\Desktop\inzynierka\aktualne\cisnienie"
base_url = "https://danepubliczne.imgw.pl/data/dane_pomiarowo_obserwacyjne/dane_meteorologiczne/dobowe/synop/"

for year in range(2014, 2024):
    year_url = f"{base_url}{year}/"
    
    response = requests.get(year_url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        
        for link in soup.find_all("a"):
            folder_name = link.get("href")
            
            if not folder_name or folder_name.startswith("../"):
                continue
            
            folder_url = f"{year_url}{folder_name}"
            response = requests.get(folder_url)
            
            if response.status_code == 200:
                folder_name_cleaned = clean_folder_name(folder_name)
                full_download_path = os.path.join(download_directory, folder_name_cleaned)
                
                with open(full_download_path, "wb") as f:
                    f.write(response.content)
            else:
                pass
    else:
        pass
```

```python
import zipfile
import os


zip_directory = r"C:\Users\48516\Desktop\inzynierka\aktualne\cisnienie"


for filename in os.listdir(zip_directory):
    if filename.endswith(".zip"):
        zip_file_path = os.path.join(zip_directory, filename)

        with zipfile.ZipFile(zip_file_path, "r") as zip_ref:
            zip_ref.extractall(zip_directory)

        print(f"Wyodrębniono: {filename}")
```

```python
import re
import pandas as pd
import os

csv_directory = r"C:\Users\48516\Desktop\inzynierka\aktualne\cisnienie"
dataframes = []

file_pattern = r"s_d_t.*\.csv"

for filename in os.listdir(csv_directory):
    if filename.endswith(".csv") and re.match(file_pattern, filename):
        csv_file_path = os.path.join(csv_directory, filename)
        encoding = 'ISO-8859-1'
        df = pd.read_csv(csv_file_path, encoding=encoding,header=None)
        dataframes.append(df)

combined_dataframe = pd.concat(dataframes, ignore_index=True)

new_column_names = {column: f"{i}" for i, column in enumerate(combined_dataframe.columns, 1)}
combined_dataframe = combined_dataframe.rename(columns=new_column_names)

filtered_dataframe = combined_dataframe[combined_dataframe.iloc[:, 1] == "KRAKÓW-OBSERWATORIUM"]

filtered_dataframe
```
