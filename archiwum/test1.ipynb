```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python


dfs = []

for month in range(1, 13):  # Assuming you have files named '2022_1.csv' to '2022_13.csv'
    filename = f'PM2.5/2022_{month}.csv'
    df = pd.read_csv(filename,header=None,names=['data','PM2.5','']).iloc[:, :-1]
    dfs.append(df)

dfs
```

```python

combined_df = pd.concat(dfs, ignore_index=True)
combined_df
```

```python

plt.figure(figsize=(12, 6)) 

x = range(len(combined_df))
y = combined_df['PM2.5']

plt.plot(x, y, label='PM2.5')
plt.xlabel('dni')
plt.ylabel('PM2.5')
plt.title('Wykres PM2.5')
plt.legend()
plt.grid()

plt.show()
```

```python
opad = []

for month in range(1, 13): 
    filename = f'temperatura/opad_2022_{month}.csv'
    df = pd.read_csv(filename,header=None)
    opad.append(df)


```

```python
opad_combined = pd.concat(opad, ignore_index=True).iloc[:,[5,7,9,11]]
opad_combined = opad_combined.rename(columns={
    opad_combined.columns[0]: 'temp',
    opad_combined.columns[1]: 'wilgotnosc',
    opad_combined.columns[2]: 'predkosc wiatru',
    opad_combined.columns[3]: 'zachmurzenie'
})
opad_combined
```
